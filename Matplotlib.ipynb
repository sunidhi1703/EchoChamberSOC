```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values and a simple quadratic function
x = np.linspace(0, 10, 100)
y = x ** 2

# Create a line plot with labels and a title
plt.plot(x, y, color='blue', linestyle='-', linewidth=2, label='y = x^2')
plt.xlabel('X value')
plt.ylabel('Y value')
plt.title('Simple Line Plot')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Generate random data
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.randint(50, 300, 50)
colors = np.random.rand(50)

# Scatter plot with size, color, alpha transparency
plt.scatter(x, y, s=sizes, c=colors, alpha=0.6, cmap='viridis')
plt.colorbar(label='Color scale')
plt.xlabel('X random')
plt.ylabel('Y random')
plt.title('Customized Scatter Plot')
plt.grid(True)
plt.show()
```

```python
# Sample categories and values
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 12]

# Create a bar chart
plt.bar(categories, values, color=['red', 'green', 'blue', 'orange'])
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Simple Bar Chart')
plt.show()
```

```python
# Generate normal distribution data
data = np.random.normal(loc=50, scale=10, size=200)

# Plot a histogram
plt.hist(data, bins=20, color='purple', edgecolor='black', alpha=0.7)
plt.xlabel('Value bins')
plt.ylabel('Frequency')
plt.title('Histogram of Normal Distribution')
plt.show()
```

```python
# Create subplots to compare different plot types
fig, axes = plt.subplots(2, 2, figsize=(10, 8), tight_layout=True)

# Line plot
axes[0, 0].plot(x, x, 'r-', label='linear')
axes[0, 0].set_title('Line Plot')
axes[0, 0].legend()

# Scatter plot
axes[0, 1].scatter(x, y, color='blue', alpha=0.5)
axes[0, 1].set_title('Scatter Plot')

# Histogram
axes[1, 0].hist(data, bins=15, color='green', edgecolor='black')
axes[1, 0].set_title('Histogram')

# Bar chart
axes[1, 1].bar(categories, values, color='cyan')
axes[1, 1].set_title('Bar Chart')

plt.show()
```
